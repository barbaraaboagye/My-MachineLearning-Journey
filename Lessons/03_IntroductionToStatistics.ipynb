### Probability
Probility is the means of knowing the outcome of an event. It has a YING -YANG (Opposite relationship) with statistics

**I LEARNT:** 
- Probability of an event : P
- Probability of an opposite event : 1 - P
- Probability of a composite event : P.P.P (Event are indepenedent)
    - Flip coin (Fair and loaded)
        - Loaded coin : The chance of seeing one side is higher than the order
         - Fair coin : Each side has a 0.5 probability (same chance)
    - Dice 
 - Conditional probability : An event happening depends on another event.
   P(TEST|DISEASE) = Probability of a disease, depends on the test results
 - Total probability
     P(TEST) =  P(TEST|DISEASE)X P(DISEASE) +  P(TEST| ¬DISEASE) X P(¬DISEASE)
