## CONTINUOUS PROBABILITY DISTRIBUTION

In continuous spaces, every outcome has a probability of 0. In continuous spaces, such as the real number line, there are an infinite number of possible outcomes. Since there are an infinite number of possible outcomes, the probability of any single outcome occurring is 0. This is because, in order to calculate the probability of an event, we divide the number of outcomes in the event by the total number of possible outcomes. Since the total number of possible outcomes is infinite, the probability of any individual outcome is 0.

Let's say we are trying to calculate the probability of a random variable X taking on a particular value x. The probability of this occurring is given by:


P(X = x) =   
$$X = \frac{\text{# of outcomes(x)}}{\text{total number of possible outcomes}}$$

Since the total number of possible outcomes in a continuous space is infinite, we can rewrite this equation as:

P(X = x) =  
$$X = \frac{x}{\text{infinity}}$$

In this case, the probability of any individual outcome is 0 because any number divided by infinity is 0. Therefore, the probability of any individual outcome in a continuous space is 0. This can be written mathematically as:

$$P(X = x) = 0$$

where x is any possible value of the random variable X

### Range probability - Spinning bottle examples

If we assume that a bottle is spinning randomly and is equally likely to land in any orientation,what is the probability of it landing 
- between 0 and 180 degrees (inclusive) ? **0.5**
    - This is because a spinning bottle has an equal chance of landing on any point on its circular surface, and the area between 0 degrees and 180 degrees on a circle is half of the total area of the circle. 
- between 260 and 290 degrees (inclusive) i.e. $260 <x \leq 290$ ? **0.0833**
    - For a circle with radius 1, Total area = 3.14 ($\pi$) square unit.
    - Area between $260 <x \leq 290$ = $\pi * 1^2 * \frac{290 -260}{360}$ = $\frac{3}{36}\pi$ square unit
    - P($260 <x \leq 290$) = $\frac{3\pi}{36\pi}$
- between 179 and 180 degrees (inclusive) ? 0.0028!
    

## Density function

For a continuous probability distribution, the y-axis represents a probability density function.
 
A density function is a mathematical function that describes the probability of an event occurring in a given range. It represent a continuous distribution.It is typically used in statistics to describe the distribution of a continuous random variable, such as the height or weight of a group of people. The density function tells you the likelihood that a given value will occur within a given range of values. For example, a density function might tell you the probability that a person chosen at random from a group will be between 5 feet and 6 feet tall.

### Real world application example

Imagine that you have a group of people, and you want to know how tall they are on average. You could take a measuring tape and measure the height of each person in the group, but that would be a lot of work. Instead, you could use a density function to estimate the average height of the group.

The density function works by dividing the range of possible heights into small intervals, like 5 feet to 5 feet and 1 inch, or 5 feet and 1 inch to 5 feet and 2 inches. Then, the density function counts how many people have a height that falls within each interval. The density function will then show you a graph with the number of people on the vertical axis and the height intervals on the horizontal axis.

The graph will show you which height intervals have the most people, and this will give you an idea of the average height of the group. For example, if the graph shows that most of the people are between 5 feet and 5 feet and 6 inches tall, then you can conclude that the average height of the group is around 5 feet and 3 inches.

So, in short, a density function is a tool that can help you quickly estimate the average value of a variable, like height or weight, for a group of people. It does this by dividing the possible values into intervals and counting how many people have a value within each interval. Then, it shows you a graph that can help you see which values are the most common.
