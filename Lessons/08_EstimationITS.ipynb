### ESTIMATOR AND AVERAGES

An estimator is a statistical model that is used to make predictions or estimates about a population based on data from a sample. An estimator typically uses a mathematical formula to make these predictions, which can be based on a variety of different factors such as the mean, median, mode, or other statistical properties of the data. Estimators are commonly used in a wide range of fields including economics, finance, and the social sciences to make predictions about trends and patterns in data. Examples of estimators include the sample mean, sample variance, and sample median.

### Maximum Likelihood Estimator (MLE)
The maximum likelihood estimator is a statistical method that uses the data from a sample to estimate the value of a population parameter. It involves finding the parameter value that maximizes the probability or likelihood of the observed data.

$$MLE = \frac{1}{N}\sum X_i$$ 
$$ N = \text{# of experiments}$$


## WEAKNESS

- The proability estimators can be a bit extreme even where there isn't enough data
 Example :
 
$$\text{The data is head} : p = 1/2 \text{(fair coin)} ; p (HEADS) = 1/2.1 = 1/2$$
   $$ \text{The data is tails};   p(HEADS) = 1/2.0 = 0$$
   
 For just one data point, the p(heads) is 0, that is the coin will always come out as tails is not really realistic. <br>
            - To solve that, fake data is added to the data

### Laplace estimator
The Laplace estimator is a statistical method that uses the data from a sample to estimate the value of a population parameter. It is often used in situations where the data is contaminated with outliers or other types of noise, and is less sensitive to the effects of these irregularities. More robust than the MLE
When there isn't much data, fake it and that's where the lapace estimator comes in.
$$\text{Laplace estimator} = \frac{1}{N + k}\sum (1 + X_i)$$
$$ N = \text{# of experiments}$$
$$ k = \text{# of outcomes}$$
