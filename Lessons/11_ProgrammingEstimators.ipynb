## Median

The median is the middle number in a list of numbers. This code calculates the median of a list containing numbers.

```python
#Complete the median function to make it return the median of a list of numbers

import math
def median(data):
    n = len(data)
    data.sort()
    #list has even number of observations
    if n % 2 == 0:
        median1 = data[n//2] #// means floor divsion - rounds the result down to the nearest integer (or whole number).
        median2 = data[n//2 - 1]
        median = (median1 + median2)/2
    else:
        median = data[n//2]
    return median


data2 = [1,4,3,2]
print(median(data1))
print(median(data2))
```

## Mode

The mode is the number that appears frequently in a list of numbers. This code calculates the mode

```python
#Complete the mode function to make it return the mode of a list of numbers

from collections import Counter

data1=[1,2,5,10,-20,5,5]

def mode(data):
    a = Counter(data)
    b = a.most_common(1)[0][0]
    return b

print (mode(data1))
```

## Variance

This is the spread of the data. This code calculates the variance. The variance formula I am using is :

 

```python
#Complete the variance function to make it return the variance of a list of numbers

#My Approach 1 
data3=[13.04, 1.32, 22.65, 17.44, 29.54, 23.22, 17.65, 10.12, 26.73, 16.43]
data1 = [3,4,5,6,7]
data2 = [17,19,18,17,19]
data4 = [85.0, 53.99, 2.67, 0.77, 96.21, 99.15, 31.27, 13.6, 66.58, 13.58, 68.48, 16.47, 0.14, 90.48, 96.31]
def mean(data):
    return sum(data)/len(data)
def variance1(data):
    a = len(data)
    l = []
    for x in data:
        c = x**2
        l.append(c)
        d = sum (l)
        b = 1/len(data)*d  -   (1/(len(data)**2)) *(sum(data)**2) #variance formula
    return b

print (variance1(data1))
print (variance1(data2))
print (variance1(data3))
print (variance1(data4))
```

```python
#Another approach - 2

def mean(data):
    return sum(data)/len(data)
def variance2(data):
    datasq = map(lambda a: a*a, data)
    b = sum(data)
    N = len(data)
    return (sum(datasq)/N) - ((b*b)/(N*N))
    
print (variance2(data1))
print (variance2(data2))
print (variance2(data3))
print (variance2(data4))
```

```python
#Another approach -3

def mean(data):
    return sum(data)/len(data)
def variance3(data):
    mu=mean(data)
    return mean([(x-mu)**2 for x in data])

print (variance3(data1))
print (variance3(data2))
print (variance3(data3))
print (variance3(data4))
```
