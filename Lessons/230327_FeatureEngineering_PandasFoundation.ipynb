## Correlation  matric

A correlation matrix is a table that displays the correlation coefficients between a set of variables or features. In statistics, correlation refers to the degree to which two variables are related to each other. A correlation coefficient is a value that indicates the strength and direction of the linear relationship between two variables. The value of a correlation coefficient ranges from -1 to +1, where -1 indicates a perfect negative correlation (as one variable increases, the other decreases), +1 indicates a perfect positive correlation (as one variable increases, the other increases), and 0 indicates no correlation (there is no linear relationship between the variables). A correlation matrix is useful when we have multiple variables and want to examine the relationships between each pair of variables


```python
# Correlation matrix

plt.matshow(df.corr())
continuous_features = df.describe().columns
plt.xticks(range(len(continuous_features)), continuous_features, rotation= 45 )
plt.yticks(range(len(continuous_features)), continuous_features, rotation= 45)
plt.colorbar()
plt.show()
```

- The `plt.matshow(df.corr())` function creates a matrix plot of the correlation coefficients between each pair of variables in the dataframe. The correlation coefficients range from -1 (perfect negative correlation) to +1 (perfect positive correlation), with 0 indicating no correlation. 

- The `continuous_features = df.describe().columns` line creates a list of the column names of the continuous variables in the dataframe. The `.describe()` method returns a summary of the descriptive statistics of the numerical columns, and `.columns` returns the column names.

- The `plt.xticks()` and `plt.yticks()` functions are used to set the labels for the x-axis and y-axis, respectively. The `range(len(continuous_features))` argument creates a sequence of integers from 0 to the length of the `continuous_features` list, which is used to set the positions of the tick marks. The `rotation= 45` argument sets the angle of the tick labels to 45 degrees for better readability.

- The `plt.colorbar()` function adds a colorbar to the plot, which shows the range of correlation coefficients and their corresponding colors.

- The `plt.show()` displays the plot on the screen.

### Preprocessing
After exploring, we can clean and preprocess our dataset

```python
# Rows with at least one NAN value

df[pd.isnull(df).any(axis =1)].head()
```

```python
# Drop rows with NAN values

df = df.dropna() #remove rows with missing values
df = df.reset_index() # reset rows index incase any number is dropped
df.head()
#df[pd.isnull(df).any(axis =1)].head() # no output since all nan have been dropped
```

```python
# Dropping multiple columns that won't be used
df = df.drop(["name", "cabin", "ticket"], axis=1) # we won't use text features for our initial basic models
df.head()
```

### Map feature values

Mapping feature values involves replacing the values of a specific column in a DataFrame with new values based on a mapping or function. This can be useful when the original values of a feature need to be transformed for better analysis or when they need to be normalized for modeling.

```python
df['sex'] = df["sex"].map( {"female" : 0, "male" : 1} ).astype(int)
df["embarked"] = df["embarked"].map( {"S" : 0, "C" : 1, "Q" :2} )
df.head()
```

### Feature engineering

This is the process of creating new features or transforming existing ones from a dataset to improve the performance of a machine learning model. It involves the extraction of relevant features from the raw data and converting them into a format that can be used by the machine learning algorithm. <br>

We're now going to use feature engineering to create a column called `family_size`. We'll first define a function called `get_family_size` that will determine the family size using the number of parents and siblings.

```python
# Lambda expressions to create new features
def get_family_size(sibsp, parch):
    family_size = sibsp + parch
    return family_size
```

Once the function is created, `lambda` can be applied to that function on each row (using the numbers of siblings and parents in each row to determine the family size for each row).

```python
df["family size"] = df[["sibsp", "parch"]].apply (lambda x:get_family_size(x["sibsp"],x["parch"]),axis =1)
df.head()
```

A new column called "family size" in a Pandas DataFrame `df` is created.

The new column is created by applying a lambda function to the columns "sibsp" and "parch" of `df`. The lambda function takes a row of "sibsp" and "parch" values as its input (specified by the axis=1 argument) and returns the output of the `get_family_size` function with "sibsp" and "parch" as arguments.

So, for each row of "sibsp" and "parch" values in `df`, the lambda function calculates the family size by passing the values to the `get_family_size function`. The resulting family size value is then assigned to the corresponding row of the new "family size" column in `df`.

```python
# Reorganize headers
df = df[["pclass", "sex", "age", "sibsp", "parch", "family size", "fare","embarked", "survived"]]
df.head()
```
