## Extract data from a table

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://www.4icu.org/top-universities-north-america/"

page = requests.get(url)

site = BeautifulSoup(page.text,"html")

print (site)
```

```python
headers = site.find_all("th")


headers_name = [header.text.strip() for header in headers]
print(headers_name)
```

```python
df = pd.DataFrame(columns = headers_name)
```

```python
column_data = site.find_all('tr')
column_data
```

```python
for row in column_data:
    row_data = row.find_all("td")
    individual_row_data = [data.text.strip() for data in row_data]
    
    if len(individual_row_data) == len(df.columns):
        length = len(df)
        df.loc[length] = individual_row_data
    else:
        # Handle rows with mismatched columns (e.g., skip or handle differently)
        print(f"Skipping row with {len(individual_row_data)} columns. Expected {len(df.columns)} columns.")


df.head
```
